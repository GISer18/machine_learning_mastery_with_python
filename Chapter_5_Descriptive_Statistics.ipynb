```python
import pandas as pd
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
df = pd.read_csv(r"./Datasets/pima-indians-diabetes.csv",names=names,header=None)
```

```python
# Peak
df.head(7) # head top 7 rows by default 5
```

```python
df.tail() # tail returns bottom rows , by default 5
```

```python
df.sample(n=6) # sample returns random number of rows
# or can ask  frac=0.25 i.e. sample 25% of total sample
```

```python
# dimensions of data
print("(rows,columns) are {}".format(df.shape))
```

```python
# dtypes
print("Dtypes of \n{}".format(df.dtypes))
```

```python
# class distribution
print("Class distribution \n",df.groupby('class').size().to_string())
```

```python
# Data Summary
df.describe()
# all terms are pretty standard
```

```python
# correlation
print("Rank correlation \n",df.corr(method='spearman'))
```

```python
# correlation
print("Pearson correlation \n",df.corr(method='pearson'))
```

```python
# skewness ( for normal distribution it is 0)
print("Skewness \n",df.skew().to_string())
```

```python
# Excess kurtosis ( for normal distribution it is 0)
print("kurtosis \n",df.kurt().to_string())
```
